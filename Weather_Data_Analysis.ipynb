```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
# Add Matplotlib inline magic command
%matplotlib inline
import os
```

```python
csv_path = os.path.join("weather_data","cities.csv")
cities_df = pd.read_csv(csv_path)
cities_df.head()
```

```python
cities_df = cities_df[["City","Country","Date","Lat","Lng","Max Temp","Humidity", "Cloudiness", "Wind Speed"]]
cities_df.head()
```

```python
import time

lat_series = cities_df['Lat']
Max_Temp_series = cities_df["Max Temp"]
humidity_series = cities_df["Humidity"]
plt.scatter(lat_series,
            Max_Temp_series,edgecolor="black", linewidths=1, marker="o",
            alpha=0.8, label="Cities")
plt.title(f"City Latitude vs. Max Temperature "+ time.strftime("%x"))
#plt.ylabel("Max Temperature (F)")
#plt.xlabel("Latitude")
#plt.grid(True)

# Save the figure.
#plt.savefig("weather_data/Fig1.png")

# Show plot.
#plt.show()

(slope, intercept, r_value, p_value, std_err) = linregress(lat_series, Max_Temp_series)
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
print(line_eq)
print(f"The p-value is: {p_value:.3f}")

regress_values = [(lat * slope + intercept) for lat in lat_series]
#print(regress_values)

# Create a scatter plot of the x and y values.
plt.scatter(lat_series,Max_Temp_series)
# Plot the regression line with the x-values and the y coordinates based on the intercept and slope.
plt.plot(lat_series,regress_values,"r")
# Annotate the text for the line equation and add its coordinates.
plt.annotate(line_eq, (10,40), fontsize=15, color="red")
plt.xlabel('Latitude')
plt.ylabel('Max Temp')
plt.show()
```

```python
plt.scatter(lat_series,
            humidity_series,edgecolor="black", linewidths=1, marker="o",
            alpha=0.8, label="Cities")
plt.title(f"City Latitude vs. Humidity "+ time.strftime("%x"))
plt.ylabel("Humidity")
plt.xlabel("Latitude")
plt.grid(True)

# Save the figure.
#plt.savefig("weather_data/Fig2.png")

# Show plot.
#plt.show()
(slope, intercept, r_value, p_value, std_err) = linregress(lat_series, humidity_series)
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
print(line_eq)
print(f"The p-value is: {p_value:.3f}")

regress_values = [(lat * slope + intercept) for lat in lat_series]
#print(regress_values)

# Create a scatter plot of the x and y values.
plt.scatter(lat_series,humidity_series)
# Plot the regression line with the x-values and the y coordinates based on the intercept and slope.
plt.plot(lat_series,regress_values,"r")
# Annotate the text for the line equation and add its coordinates.
plt.annotate(line_eq, (10,40), fontsize=15, color="red")
plt.xlabel('Latitude')
plt.ylabel('Humidity')
plt.show()

```

```python
from scipy.stats import linregress

# Create an equal number of latitudes and temperatures.
lats = [42.5, 43.9, 8.1, 36.8, 79.9, 69.1, 25.7, 15.3, 12.7, 64.5]
temps = [80.5, 75.3, 90.9, 90.0, 40.4, 62.3, 85.4, 79.6, 72.5, 72.0]

# Perform linear regression.
(slope, intercept, r_value, p_value, std_err) = linregress(lats, temps)
# Get the equation of the line.
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
print(line_eq)
print(f"The p-value is: {p_value:.3f}")
```

```python
regress_values = [(lat * slope + intercept) for lat in lats]
print(regress_values)
```

```python
# Import Matplotlib.
import matplotlib.pyplot as plt
# Create a scatter plot of the x and y values.
plt.scatter(lats,temps)
# Plot the regression line with the x-values and the y coordinates based on the intercept and slope.
plt.plot(lats,regress_values,"r")
# Annotate the text for the line equation and add its coordinates.
plt.annotate(line_eq, (10,40), fontsize=15, color="red")
plt.xlabel('Latitude')
plt.ylabel('Temp')
plt.show()
```

```python
north_hemisphere_df = cities_df.loc[(cities_df["Lat"] >= 0)]
south_hemisphere_df = cities_df.loc[(cities_df["Lat"] <= 0)]
north_hemisphere_df.head()
```

```python
north_lat_series = north_hemisphere_df["Lat"]
north_temp_series = north_hemisphere_df["Max Temp"]
(slope, intercept, r_value, p_value, std_err) = linregress(north_lat_series, north_temp_series)

# Get the equation of the line.
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
print(line_eq)
print(f"The p-value is: {p_value:.3f}")

regress_values = [(lat * slope + intercept) for lat in north_lat_series]
#print(regress_values)

plt.title(f"City Latitude vs. Max Temperature in North Hemisphere "+ time.strftime("%x"))
plt.scatter(north_lat_series,north_temp_series)
# Plot the regression line with the x-values and the y coordinates based on the intercept and slope.
plt.plot(north_lat_series,regress_values,"r")
# Annotate the text for the line equation and add its coordinates.
plt.annotate(line_eq, (10,40), fontsize=15, color="red")
plt.xlabel('Latitude')
plt.ylabel('Max Temp')
plt.show()
```
